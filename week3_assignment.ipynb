```python
import numpy as np
```

# 1. 0~9까지의 값으로 넘파이 1차원 배열을 만들고, 이 배열을 거꾸로 나열하라.

```python
arr = np.arange(10)
print(f"원래 배열 : {arr}")

sort_arr = np.flip(arr)
print(f"내림차순 정렬 배열 : {sort_arr}")
```

# 2. 0~8까지의 값을 가지고 3x3 행렬과 9x1행렬을 생성하라.

```python
arr = np.arange(9)
arr_3by3 = arr.reshape(3,3)
arr_9by1 = arr.reshape(9,1)

print(f"원래 배열 \n{arr}", end = '\n\n')
print(f"3x3 배열 \n{arr_3by3}", end = '\n\n')
print(f"9x1 배열 \n{arr_9by1}", end = '\n\n')
```

# 3. 배열의 테두리에 1, 내부에 0을 가진 10x10 크기의 행렬을 생성하라.

```python
arr = np.zeros((10,10))
arr[[0, -1], :] = 1
arr[:, [0, -1]] = 1

print(arr)
```

# 4. 5x5 행렬을 만들어 체스보드 패턴으로 채워보라.(슬라이싱 이용)

```python
arr = np.zeros((5,5))
arr[1::2, ::2] = 1
arr[::2, 1::2] = 1
print(arr)
```

# 5. 넘파이를 사용해서 0~9까지의 값을 가지는 벡터를 만들고 5에서 8 사이의 숫자 부호를 반전시켜보라


```python
v = np.arange(10)
v[5:9] *= -1
print(v, v.shape)
```

# 6. 주어진 두 배열의 내적을 계산하라.

```python
x = np.array([[4, 5], [7, 10]])
y = np.array([[2, 3], [1, 5]])

result = x @ y
print(result)
```

# 7. 시그모이드(Sigmoid) 함수의 1차 미분 함수를 구하고, Sigmoid 함수와 미분 함수를 같은 화면에 그려보라.

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
# 임의의 z값 생성
z = np.linspace(-10, 10, 100)

# 시그모이드 정의
sigmoid = 1 / (1 + np.exp(-z))

# 시그모이드 미분 정의
def sigmoid_diff(z):
    sigmoid =  1 / (1 + np.exp(-z))
    return sigmoid * (1 - sigmoid)

plt.figure(figsize=(14, 5))

# 시각화
plt.subplot(121)
plt.plot(z, sigmoid, color='k')
plt.xlabel('z')
plt.ylabel('시그모이드')
plt.title('시그모이드 그래프')

plt.subplot(122)
plt.plot(z, sigmoid_diff(z), color='k')
plt.xlabel('z')
plt.ylabel('시그모이드 미분값')
plt.title('시그모이드 미분 그래프')

plt.show()
```
