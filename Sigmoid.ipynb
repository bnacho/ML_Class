```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

# 오즈(Odds)
- 어떠한 사건이 일어날 확률과 일어나지 않을 확률의 비


```python
# 임의의 확률값 생성
p = np.linspace(0.01, 0.99, 100)

# 오즈 정의
odds = p / (1-p)

# 확률에 따른 오즈값 시각화
plt.figure(figsize=(10, 6))
plt.plot(p, odds, color='k')
plt.xlabel('확률')
plt.ylabel('오즈')
plt.title('확률과 오즈 그래프')
plt.show()
```

# 로짓(Logit)
- 오즈는 범위의 제약이 있다.(0부터 무한대까지)
- 범위의 제약을 없애기 위해 오즈에 자연로그를 취해준 것이 로짓이다.
- 

```python
# 임의의 확률값 생성
p = np.linspace(0.01, 0.99, 100)

# 로짓 정의
logit = np.log(p / (1 - p))

# 확률에 따른 로짓값 시각화
plt.figure(figsize=(10, 6))
plt.plot(p, logit, color='k')
plt.xlabel('확률')
plt.ylabel('로짓')
plt.title('확률과 로짓 그래프')
plt.show()
```

# 시그모이드 함수(Sigmoid Function)
- 위 로지스틱 회귀 모델 식의 꼴을 시그모이드 함수라고 한다.
- 0일 때 0.5를 지난다.
- 

```python
# 임의의 z값 생성
z = np.linspace(-10, 10, 100)

# 시그모이드 정의
sigmoid = 1 / (1 + np.exp(-z))

# z에 따른 시그모이드값 시각화
plt.figure(figsize=(10, 6))
plt.plot(z, sigmoid, color='k')
plt.xlabel('z')
plt.ylabel('시그모이드')
plt.title('시그모이드 그래프')
plt.show()
```

# 시그모이드 미분(Sigmoid Differential)
- 몫의 미분법을 사용하고 잘 정리하면 다음과 같은 결과가 나온다.
- 

```python
# 시그모이드 미분 함수 정의
def sigmoid_diff(z):
    sigmoid =  1 / (1 + np.exp(-z))
    return sigmoid * (1 - sigmoid)
    
# 임의의 z값 생성
z = np.linspace(-10, 10, 100)

# z에 따른 시그모이드값 시각화
plt.figure(figsize=(10, 6))
plt.plot(z, sigmoid_diff(z), color='k')
plt.xlabel('z')
plt.ylabel('시그모이드 미분값')
plt.title('시그모이드 미분 그래프')
plt.show()
```
